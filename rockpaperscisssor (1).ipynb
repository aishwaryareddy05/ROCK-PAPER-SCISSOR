```python
from tkinter import *
from PIL import Image,ImageTk
import random
root=Tk()
root.title("ROCK-PAPER-SCISSOR")
root.configure(background="pink")
rock_user =ImageTk.PhotoImage(Image.open("user_rock-min.png"))
rock_comp =ImageTk.PhotoImage(Image.open("comp_rock-min.png"))
scissor_user =ImageTk.PhotoImage(Image.open("user_scissor-min.png"))
scissor_comp =ImageTk.PhotoImage(Image.open("comp_scissor-min.png"))
paper_user =ImageTk.PhotoImage(Image.open("user_paper-min.png"))
paper_comp =ImageTk.PhotoImage(Image.open("comp_paper-min.png"))
user_label=Label(root,image=rock_user,bg="pink")
comp_label=Label(root,image=rock_comp,bg="pink")
comp_label.grid(row=1,column=4)
user_label.grid(row=1,column=0)
#indicators
user_ind=Label(root,font=100,text="USER",bg="pink")
comp_ind=Label(root,font=100,text="COMPUTER",bg="pink")
user_ind.grid(row=0,column=1)
comp_ind.grid(row=0,column=3)
#buttons
rock=Button(root,width=20,height=2,text="ROCK",bg="#64b5f6",fg="white",command=lambda:updChoice("rock")).grid(row=2,column=1)
scissor=Button(root,width=20,height=2,text="SCISSOR",bg="#fdb900",fg="white",command=lambda:updChoice("scissor")).grid(row=2,column=2)
paper=Button(root,width=20,height=2,text="PAPER",bg="#d32f2f",fg="white",command=lambda:updChoice("paper")).grid(row=2,column=3)
#messages
msg=Label(root,font=2000,bg="pink",fg="white")
msg.grid(row=1,column=2)
#update choices
choices=["rock","scissor","paper"]
def updChoice(x):
    #forcomputer
    compch=random.choice(choices)
    if compch=="rock":
        comp_label.configure(image=rock_comp)
    elif compch=="paper":
         comp_label.configure(image=paper_comp)
    else:
         comp_label.configure(image=scissor_comp)
    
    #for useer
    if x=="rock":
        user_label.configure(image=rock_user)
    elif x=="paper":
         user_label.configure(image=paper_user)
    else:
         user_label.configure(image=scissor_user)
    checkWin(x,compch)
#update message
def updMsg(x):
    msg["text"]=x
def checkWin(x,compch):
    if x==compch:
        updMsg("IT'S A TIE!")
    elif x=="rock":
        if compch=="scissor":
            updMsg("YOU WIN")
        else:
            updMsg("YOU LOSE")
    elif x=="scissor":
        if compch=="rock":
            updMsg("YOU LOSE")
        else:
            updMsg("YOU WIN")
    elif x=="paper":
        if compch=="scissor":
            updMsg("YOU LOSE")
        else:
             updMsg("YOU WIN")
            
    

    
        

root.mainloop()
```
